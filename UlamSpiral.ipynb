```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def is_prime_fast(n):
    '''
    Checks whether the argument n is a prime number.
    Uses a brute force search for factors between 1 and n.
    '''
    j = 2
    root_n = np.sqrt(n)
    while j <= root_n:  # j will proceed through the list of numbers 2,3,... up to sqrt(n).
        if n%j == 0:  # is n divisible by j?
            #print "%d is a factor of %d."%(j,n)
            return False
        j = j + 1  # There's a Python abbreviation for this:  j += 1.
    return True
```

```python
is_prime_fast(105)
```

```python
size = 1000
x = 0
y = 0
d = {}
d[(0,0)] = 0.0
n = 0
sign = 1
for stepsize in range(1, size+1):
    for step in range(1, stepsize+1):
        n = n + 1
        x = x + sign
        if is_prime_fast(n):
            d[(x,y)] = 1.0
        else:
            d[(x,y)] = 0.0
    for step in range(1, stepsize+1):
        n = n + 1
        y = y + sign
        if is_prime_fast(n):
            d[(x,y)] = 1.0
        else:
            d[(x,y)] = 0.0
    sign = -1 * sign
```

```python
arr = np.zeros((size+1,size+1))
for k,v in d.iteritems():
    (x,y) = k
    arr[x+size/2,y+size/2] = v
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 15)
ax.imshow(arr,  cmap='inferno', aspect='auto')
plt.show()
```
